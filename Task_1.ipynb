```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the dataset
# Seaborn has it built-in, avoiding the need to download external CSVs.
df = sns.load_dataset('iris')

# 2. Inspecting the data
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst 5 rows:")
print(df.head())

# 3. Summary Statistics
print("\nDataset Info:")
df.info()
print("\nSummary Statistics:")
print(df.describe())

# 4. Visualization
sns.set_theme(style="whitegrid")

# Scatter plot: Using a pairplot to see ALL relationships at once
sns.pairplot(df, hue="species")
plt.suptitle("Feature Relationships by Species", y=1.02)
plt.show()

# Histograms: Showing value distributions
df.drop('species', axis=1).hist(bins=20, figsize=(10, 8))
plt.suptitle("Feature Distributions")
plt.show()

# Box plots: Identifying outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('species', axis=1))
plt.title("Feature Box Plots (Outlier Detection)")
plt.show()

```
