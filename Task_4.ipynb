```python
from openai import OpenAI

# Initialize the OpenAI client with your API key
client = OpenAI(api_key="YOUR_API_KEY")

# System prompt using RTCF framework: (Role, Task, Constraints, Format)
SYSTEM_PROMPT = """
You are a friendly and knowledgeable health assistant with expertise in general medical information.

Your task is to answer general health questions clearly and simply, help users understand symptoms, medications, and wellness topics,
and maintain a natural back-and-forth conversation.

Never diagnose conditions or prescribe medications.
Never provide advice that replaces a licensed doctor.
If a question involves serious symptoms, mental health crises, or emergencies,
always redirect the user to seek professional help immediately.
Avoid technical jargon unless the user asks for it.

Keep responses concise and easy to understand.
End every response with a one-line reminder to consult a doctor for personal medical advice.
Use plain conversational language, no bullet points unless listing multiple items.
"""

# conversation_history stores the full chat so the model remembers context
# It always starts with the system prompt at index 0
conversation_history = [{"role": "system", "content": SYSTEM_PROMPT}]

# Maximum number of messages to keep in memory (excluding system prompt)
# 10 messages = 5 user turns + 5 assistant replies
MAX_HISTORY = 10

def chat(user_message):
    conversation_history.append({"role": "user", "content": user_message})

    while len(conversation_history) > MAX_HISTORY + 1:
        conversation_history.pop(1)  # remove oldest user message
        conversation_history.pop(1)  # remove its assistant reply

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=conversation_history
    )

    reply = response.choices[0].message.content
    conversation_history.append({"role": "assistant", "content": reply})

    return reply

# Run the chatbot & keeps the chatbot running until user types 'quit'
print("Health Assistant ready. Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("Goodbye. Stay healthy!")
        break
    response = chat(user_input)
    print(f"\nAssistant: {response}\n")
```
